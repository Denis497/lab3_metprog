```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('prng_timing.csv')

plt.figure()
plt.plot(df['N'], df['LCG'], label='LCG')
plt.plot(df['N'], df['Xorshift'], label='Xorshift32')
plt.plot(df['N'], df['Xorshift128'], label='Xorshift128')
plt.plot(df['N'], df['MT19937'], label='std::mt19937')
plt.xlabel('  (N)')
plt.ylabel('  ()')
plt.title('     ')
plt.legend()
plt.tight_layout()
plt.show()
```
